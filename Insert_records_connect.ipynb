```python
import pandas as pd     #(version 1.0.0)
import plotly           #(version 4.5.0)
import plotly.express as px

import dash             #(version 1.8.0)
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output, State

app = dash.Dash(__name__)

#---------------------------------------------------------------

df = pd.read_csv("food_coded.csv") 
df.info()
```

```python
# create a list of our conditions :gender
conditions1 = [(df['gender'] == 1),(df['gender'] == 2),]
values1 = ['Female', 'Male']
df['gender_mean'] = np.select(conditions1, values1)

# create a list of our conditions :calories_day
conditions2 = [(df['calories_day'] == 1),(df['calories_day'] == 2),(df['calories_day'] == 3),(df['calories_day'] == 4),(df['calories_day'] == 0)]
values2 = ['ฉันไม่รู้ว่าฉันควรกินกี่แคล', 'มันไม่สำคัญเลย','มีความสำคัญปานกลาง', 'มันสำคัญมาก',"none"]
df['calories_day_mean'] = np.select(conditions2, values2)


# create a list of our conditions :cook
conditions3 = [(df['cook'] == 1),(df['calories_day'] == 2),(df['calories_day'] == 3),(df['calories_day'] == 4),(df['calories_day'] == 5)]
values3 = ['ทุกวัน', 'สองสามครั้งต่อสัปดาห์','เมื่อไหร่ก็ได้ แต่ก็ไม่บ่อยนัก', 'ฉันช่วยเพียงเล็กน้อยในช่วงวันหยุด','ไม่เลย ฉันไม่รู้ทางในครัวจริง']
df['cook_mean'] = np.select(conditions3, values3)


# create a list of our conditions :comfort_food_reasons_coded_1
conditions4 = [(df['comfort_food_reasons_coded'] == 1),(df['comfort_food_reasons_coded'] == 2),(df['comfort_food_reasons_coded'] == 3),(df['comfort_food_reasons_coded'] == 4),(df['comfort_food_reasons_coded'] == 5),(df['comfort_food_reasons_coded'] == 6),(df['comfort_food_reasons_coded'] == 7),(df['comfort_food_reasons_coded'] == 8),(df['comfort_food_reasons_coded'] == 9)]
values4 = ['stress', 'boredom','depression/sadness', 'hunger','laziness','cold weather', 'happiness','watching tv', 'none']
df['comfort_food_reasons_coded_1_mean'] = np.select(conditions4, values4)

df
```

```python
#graph 1 Top 20 Members of Count Message 

groupby_gender=(df.groupby(['gender_mean']).count()[['ids']].sort_values('gender_mean', ascending=False))
groupby_gender

import plotly.graph_objects as go
labels = ['Male','Female']
values = [49,76]
```

```python
#graph 2: The proportion of students assessed the importance of daily calorie consumption

df_calories_day=(df.groupby(['calories_day_mean']).count()[['ids']].sort_values('calories_day_mean', ascending=False))

df_calories_day = df_calories_day.dropna()
```

```python
#graph 4 คุณทำอาหารบ่อยแค่ไหน ?

cook_freq = (df.groupby(['cook_mean']).count()[['ids']].sort_values('cook_mean', ascending=False))


#graph 5 คุณกิน comfort food เมื่อมีอารมณ์แบบใด
com_food_reason = (df.groupby(['comfort_food_reasons_coded1']).count()[['comfort_food_reasons']].sort_values('comfort_food_reasons', ascending=False))
#top_comment = Member_comment.nlargest(20, 'Message')
com_food_reason
```

```python
#graph 5 คุณกิน comfort food เมื่อมีอารมณ์แบบใด

com_food_reason = (df.groupby(['comfort_food_reasons_coded1']).count()[['comfort_food_reasons']].sort_values('comfort_food_reasons', ascending=False))
#top_comment = Member_comment.nlargest(20, 'Message')
com_food_reason


#graph 6 คุณมีพฤติกรรมกินอาหารอย่างไร ?
diet_current = (df.groupby(['diet_current_coded']).count()[['diet_current']].sort_values('diet_current_coded', ascending=False))
#top_comment = Member_comment.nlargest(20, 'Message')
diet_current
```
